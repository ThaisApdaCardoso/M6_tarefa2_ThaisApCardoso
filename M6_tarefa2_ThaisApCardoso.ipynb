####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import matplotlib.pyplot as plt

arquivo = 'sinasc_RO_2019.csv'
df = pd.read_csv(arquivo)

#Criando o boxplot da idade da mãe
plt.figure(figsize=(8, 6))
plt.boxplot(df['IDADEMAE'].dropna(), vert=False)
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.grid(True)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
#Convertendo o peso do bebê de gramas para quilos

df['PESO_KG'] = df['PESO'] / 1000

#Criando o boxplot do peso do bebê

plt.figure(figsize=(8, 6))
plt.boxplot(df['PESO_KG'].dropna(), vert=False)
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê (kg)')
plt.grid(True)
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
#histograma da idade da mãe

plt.figure(figsize=(8, 6))
plt.hist(df['IDADEMAE'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
#histograma do peso do bebê (quilos)

plt.figure(figsize=(8, 6))
plt.hist(df['PESO_KG'].dropna(), bins=20, color='lightcoral', edgecolor='black')
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê (kg)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
```
